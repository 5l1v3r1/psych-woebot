## Conclusion

- Results **EXPECTED** were:
    - Number of people with **severe depression** would **decrease** with resepect to to PHS-9 scores
    - Number of people who are **not at all depressed** would **go up** with resepect to to PHS-9 scores
    - It was also expected that interaction with the chatbot would enable the participants to be **less affected by the factors listed in PHS-9**
    - We also expected the **PANAS-15 Positive Scores to go up** and **Negative Scores to come down** 
    
    
    
- Results **OBTAINED** were:
    - Although the number of people who are **not at all depressed** increased, the number of people with **severe depression** suffered **an increase followed by a decrease**, with respect to PHS-9 scores
    - The number of people who **feel affected** by the factors listed in **PHS-9** went up through the survey
    - The **PANAS-15 Positive Score went down whereas the Negative Score went up** through the experiment, **contrary to expecations**
    
    
    
- **Possible Reasons**:
    - The participants did not follow the experiment instructions properly
    - The Volunteer count was so low that biases of individual volunteers swayed the results
    - The conduction time of the survey was very small and this is by no means a rogorous statistical analysis, hence the figures and our expectations may not be in the correct direction
    
Apart from this we also asked the volunteers an optional question about their experience with "Woebot". Out of the 8 volunteers who answered that question, the responses were mostly positive. The only major flaw detected was the inability of woebot to handle long open-ended dialogues.
